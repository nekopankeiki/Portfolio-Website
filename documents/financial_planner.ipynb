```python
# install dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import numpy as np
import random

# create simulation data according to guidelines given
data = {'Timestamp':['31/03/2024 9:00', '31/03/2024 12:00', '31/03/2024 18:00', '01/04/2024 14:00', '01/04/2024 15:30', '02/04/2024 9:00', '02/04/2024 11:00'], 
        'Type of expenses':['Transport', 'Food', 'Food', 'Food', 'Transport', 'Lifestyle', 'Miscellaneous'],
       'Expense value':[1.50, 6.00, 10.00, 9.00, 1.50, 15.00, 5.00]}
data['Timestamp'] = pd.to_datetime(data['Timestamp'], dayfirst=True)

finance_tracker_df = pd.DataFrame(data)
typedict = {1:'Food', 2:'Transport', 3:'Lifestyle', 4:'Miscellaneous'}
for i in range(45):
    finance_tracker_df.loc[len(finance_tracker_df)] = {
        'Timestamp': pd.Timestamp.now(),
        'Type of expenses': typedict[random.randint(1, 4)],
        'Expense value': random.randint(1, 15)}

# format the dataframe
finance_tracker_df.sort_values(by='Timestamp', inplace=True)
total_expenditure = finance_tracker_df['Expense value'].cumsum()
finance_tracker_df['Total expenditure'] = total_expenditure
total_savings = 1000 - finance_tracker_df['Total expenditure']
finance_tracker_df['Total savings'] = total_savings
display(finance_tracker_df)

# save the data as a new csv
finance_tracker_df.to_csv("finance_tracker.csv", index=False)

# add function to modify csv based on user input 
def financemodify():
    global finance_tracker_df 
    global total_expenditure
    global total_savings
    while True:
        user_input = input("Is there anything you wish to modify? (Yes/No) ")
        if user_input == "Yes":
            add_or_delete = input("Add or delete?")
            if add_or_delete== "Add":
                type_of_expense = input("Type of expense?")
                value_of_expense = int(input("Value of expense?"))
                total_expenditure += value_of_expense
                total_savings = 1000 - total_expenditure
                current_dateTime = datetime.now()
                finance_tracker_df.loc[len(finance_tracker_df.index)] = [current_dateTime, type_of_expense, value_of_expense, total_expenditure.iloc[-1], total_savings.iloc[-1]]
                display(finance_tracker_df)
            elif add_or_delete == "Delete":
                row_to_delete = int(input("Indicate which row you wish to delete "))
                finance_tracker_df = finance_tracker_df.drop(row_to_delete, axis='index')
                new_total_expenditure = finance_tracker_df['Expense value'].cumsum()
                finance_tracker_df['Total expenditure'] = new_total_expenditure
                new_total_savings = 1000 - finance_tracker_df['Total expenditure']
                finance_tracker_df['Total savings'] = new_total_savings
                display(finance_tracker_df)
            else:
                print("You cannot modify that!")
                continue
            break
        else:
            pass
        break
        
financemodify()

# plot pie chart showing expenditure
expense_tracker = finance_tracker_df.iloc[:, 1:3] # remove datetime columns 
finance_pie_chart = expense_tracker.groupby(['Type of expenses']).sum().plot(kind='pie', y='Expense value', 
                                                                   autopct='%1.0f%%').legend(bbox_to_anchor=(1.5, 1))

# plot line graph showing change in spending habit over time
finance_line_graph = finance_tracker_df.plot(x='Timestamp', y='Expense value')

# plot comparative bar plot
labels = finance_tracker_df['Timestamp'].tolist()
index = np.arange(len(labels)) 
bar_width = 0.1

fig, ax = plt.subplots(figsize=(20,5))
expenditure = ax.bar(index, finance_tracker_df['Total expenditure'], bar_width,
                label="Expenditure")

savings = ax.bar(index+bar_width, finance_tracker_df['Total savings'],
                 bar_width, label="Savings")

ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=90)
ax.set_xlabel('Time')
ax.set_ylabel('Amount')
ax.legend()

```
